```python
# Selecting the numeric column "Amount ($)" for analysis
amount_column = data_1['Amount ($)']

# Calculating the mean, median, and mode
mean_amount = amount_column.mean()
median_amount = amount_column.median()
mode_amount = amount_column.mode().iloc[0]  

# Print the results
print(f"Mean: {mean_amount:.2f}")
print(f"Median: {median_amount:.2f}")
print(f"Mode: {mode_amount:.2f}")
```

```python
import csv
from statistics import mean, median, mode


file = 'New_York_City_Council_Discretionary_Funding.csv'
amounts = []

with open(file, newline='', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        value = row['Amount ($)']
        if value.replace('.', '', 1).isdigit():  
            amounts.append(float(value))


mean_amount = round(mean(amounts), 2)
median_amount = round(median(amounts), 2)
mode_amount = round(mode(amounts), 2)


print(f"Mean: {mean_amount}")
print(f"Median: {median_amount}")
print(f"Mode: {mode_amount}")
```

```python
import csv
import pandas as pd

# Load the dataset
data = pd.read_csv('New_York_City_Council_Discretionary_Funding.csv')

# Aggregate "Amount ($)" by "Fiscal Year"
aggregated_data = data.groupby("Fiscal Year")["Amount ($)"].sum()

# Normalize data for visualization
max_value = aggregated_data.max()
min_value = aggregated_data.min()
chart_data = aggregated_data.reset_index()

# Generate a vertical bar chart
for _, row in chart_data.iterrows():
    year = row['Fiscal Year']
    total_amount = row['Amount ($)']
    # Scale bar length to max 50 characters
    bar_length = int((total_amount - min_value) / (max_value - min_value) * 50)
    bar = "█" * bar_length
    print(f"{year}: {bar} {total_amount:.2f}")
```
